```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

def timeConverter(df):
    format = '%H:%M'
    return datetime.strptime(df,format)

def timeConverter2(df):
    if (df.seconds/60 > 1000) :
        return (1440-(df.seconds/60))*-1
    elif (df.seconds/60 < 1000) :
        return df.seconds/60

#have to use return of array for missing keys because cant reference unknown key in dictionary
def airlineFunction(al):
    if al[2:3].isdigit():
        if airlinekey.get(al[0:2],"Unknown")=="Unknown":
            return "Unknown"
        return airlinekey[al[0:2]]["airline"]
    elif al[2:3].isalpha():
        if airlinekey.get(al[0:3],"Unknown")=="Unknown":
            return "Unknown"
        return airlinekey[al[0:3]]["airline"] 
    
def handlerFunction(al):
    if al[2:3].isdigit():
        if airlinekey.get(al[0:2],"Unknown")=="Unknown":
            return "Unknown"
        return airlinekey[al[0:2]]["handler"]
    elif al[2:3].isalpha():
        if airlinekey.get(al[0:3],"Unknown")=="Unknown":
            return "Unknown"
        return airlinekey[al[0:3]]["handler"]
    
def getHour(ft):
    return ft[0:2]
 
def calcThreshold(tt):
    if tt>=15:
        return 1
    elif tt<15:
        return 0
    
    
airlinekey = {
		"FR": {"airline" : "Airline 1", "handler" : "Handler 1"},
		"EI": {"airline" : "Airline 2", "handler" : "Handler 2"},
		"0B": {"airline" : "Airline 3", "handler" : "Handler 3"},
		"3V": {"airline" : "Airline 4", "handler" : "Handler 4"},
		"5X": {"airline" : "Airline 5", "handler" : "Handler 4"},
		"9U": {"airline" : "Airline 6", "handler" : "Handler 4"},
		"AA": {"airline" : "Airline 7", "handler" : "Handler 3"},	
		"AF": {"airline" : "Airline 8", "handler" : "Handler 4"},
		"AY": {"airline" : "Airline 9", "handler" : "Handler 3"},
		"BA4": {"airline" : "Airline 10", "handler" : "Handler 4"},	
		"BA": {"airline" : "Airline 11", "handler" : "Handler 4"},	
		"BBD": {"airline" : "Airline 12", "handler" : "Handler 4"},
		"BE": {"airline" : "Airline 13", "handler" : "Handler 4"},	
		"CX": {"airline" : "Airline 14", "handler" : "Handler 4"},
		"D8": {"airline" : "Airline 15", "handler" : "Handler 4"},	 
		"DL": {"airline" : "Airline 16", "handler" : "Handler 3"},	
		"DY": {"airline" : "Airline 17", "handler" : "Handler 4"},
		"EI3": {"airline" : "Airline 18", "handler" : "Handler 4"},
		"EK": {"airline" : "Airline 19", "handler" : "Handler 3"},
		"ENT": {"airline" : "Airline 20", "handler" : "Handler 3"},
		"ET": {"airline" : "Airline 21", "handler" : "Handler 3"},
		"EY": {"airline" : "Airline 22", "handler" : "Handler 3"},
		"EW": {"airline" : "Airline 23", "handler" : "Handler 3"},
		"FIA": {"airline" : "Airline 24", "handler" : "Handler 3"},
		"FX": {"airline" : "Airline 25", "handler" : "Handler 3"},
		"HU": {"airline" : "Airline 26", "handler" : "Handler 4"},
		"I2": {"airline" : "Airline 27", "handler" : "Handler 3"},
		"KL": {"airline" : "Airline 28", "handler" : "Handler 4"},
		"LG": {"airline" : "Airline 29", "handler" : "Handler 4"},
		"LH": {"airline" : "Airline 30", "handler" : "Handler 4"},
		"LM": {"airline" : "Airline 31", "handler" : "Handler 4"},
		"LX": {"airline" : "Airline 32", "handler" : "Handler 4"},
		"OE": {"airline" : "Airline 33", "handler" : "Handler 1"},
		"OU": {"airline" : "Airline 34", "handler" : "Handler 3"},
		"QR": {"airline" : "Airline 35", "handler" : "Handler 4"},
		"QY": {"airline" : "Airline 36", "handler" : "Handler 3"},
		"REV": {"airline" : "Airline 37", "handler" : "Handler 3"},
		"SIG": {"airline" : "Airline 38", "handler" : "Handler 5"},
		"SK": {"airline" : "Airline 39", "handler" : "Handler 4"},
		"SU": {"airline" : "Airline 40", "handler" : "Handler 4"},
		"TK": {"airline" : "Airline 41", "handler" : "Handler 3"},
		"TO": {"airline" : "Airline 42", "handler" : "Handler 4"},
		"TOM": {"airline" : "Airline 43", "handler" : "Handler 3"},
		"UA": {"airline" : "Airline 44", "handler" : "Handler 3"},
		"UVA": {"airline" : "Airline 45", "handler" : "Handler 6"},
		"VY": {"airline" : "Airline 46", "handler" : "Handler 3"},
		"TP": {"airline" : "Airline 47", "handler" : "Handler 3"},
		"TS": {"airline" : "Airline 48 ", "handler" : "Handler 4"},
		"WS": {"airline" : "Airline 49", "handler" : "Handler 4"},
		"XM": {"airline" : "Airline 50", "handler" : "Handler 3"},
		"A3": {"airline" : "Airline 51", "handler" : "Handler 3"},
		"AC": {"airline" : "Airline 52", "handler" : "Handler 3"},
		"BT": {"airline" : "Airline 53", "handler" : "Handler 4"},
		"DW": {"airline" : "Airline 54", "handler" : "Handler 3"},
		"FI": {"airline" : "Airline 55", "handler" : "Handler 4"},
		"NJE": {"airline" : "Airline 56", "handler" : "Handler 6"},
		"AP": {"airline" : "Airline 57", "handler" : "Handler 3"},
		"ENZ": {"airline" : "Airline 58", "handler" : "Handler 3"},
        "PVT": {"airline" : "Airline 59", "handler" : "None"},
        "RES": {"airline" : "Helicopter", "handler" : "None"},
        "BID": {"airline" : "Airline 60", "handler" : "Handler 3"},
        "BZE": {"airline" : "General Aviation", "handler" : "Handler 6"},
        "DHK": {"airline" : "Airline 61", "handler" : "Handler 3"}}

flights19 = pd.read_csv('FlightInfoAug2019.csv')
flights19['year']="2019"
flights20 = pd.read_csv('FlightInfoAug2020.csv')
flights20['year']="2020"
flights21 = pd.read_csv('FlightInfoAug2021.csv')
flights21['year']="2021"
flights = pd.concat([flights19, flights20, flights21],ignore_index=True )
flights = flights.sort_values(by=['Date - Scheduled Date DateName'])
flights["index"]= flights["Date - Scheduled Date DateName"]+ flights["Flights - Attributes Flight Number"]
flights = flights.set_index('index')
flights = flights.drop_duplicates(subset = ['Date - Scheduled Date DateName', 'Flights - Attributes Flight Number'])
flights = flights[flights['Time - Block Time Time'] != "Unknown Time"] 
flights = flights[flights['Flight Status'] != "Cancelled"]
flightsSub = flights[:][['year','Time - Scheduled Time Time','Time - Block Time Time','Flights - Attributes Flight Number','Date - Scheduled Date DateName']]
flightsSub["landingTime"]= flightsSub['Time - Scheduled Time Time'].apply(timeConverter)
flightsSub["blockTime"]= flightsSub['Time - Block Time Time'].apply(timeConverter)
flightsSub["delta"] = (flightsSub["blockTime"] - flightsSub["landingTime"])
flightsSub["taxiTime"] = (flightsSub["delta"].apply(timeConverter2))
flightsSub["airline"] = (flightsSub["Flights - Attributes Flight Number"].apply(airlineFunction))
flightsSub["handler"] = (flightsSub["Flights - Attributes Flight Number"].apply(handlerFunction))
flightsSub["hour"] = flightsSub['Time - Scheduled Time Time'].apply(getHour)
#for counting number of rows by any dimension for plotting
flightsSub["rowCount"] = 1;
flightsSub["investigationThreshold"] = flightsSub["taxiTime"].apply(calcThreshold)

fig = plt.figure()
fig.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=2, wspace=None, hspace=None)
ax1 = fig.add_subplot(521)
ax2 = fig.add_subplot(522)

#flights per airline
ax1.set_xlabel('Airline')
ax1.set_ylabel('Number of Flights')
ax1.set_title('Fig1: Total Flights (All Years) by Airline')
flightsSub["airline"].value_counts().plot(kind='bar',log = True,  figsize=(30,20), ax=ax1)

#flights per handler
ax2.set_xlabel('Handler')
ax2.set_ylabel('Number of Flights')
ax2.set_title('Fig 2: Total Flights (All Years) by Handler')
flightsSub["handler"].value_counts().plot(kind='bar',log = True,  figsize=(30,20), ax=ax2)

flightsSub = flightsSub.sort_values(by=['hour'])

#total flights per hour
ax3 = fig.add_subplot(523)
ax3.set_xlabel('Hour')
ax3.set_ylabel('Number of Flights')
ax3.set_title('Fig 3: Total Flights (All Years) by Hour')
flightsSub.groupby("hour")['rowCount'].agg(['sum']).plot(kind='bar',  figsize=(30,20), ax = ax3)

#mean taxi time per hour
ax4 = fig.add_subplot(524)
ax4.set_xlabel('Hour')
ax4.set_ylabel('Average Taxi Time (Minutes)')
ax4.set_title('Fig 4: Mean Taxi Time per Hour')              
flightsSub.groupby('hour')['taxiTime'].agg(['mean']).plot(kind='bar', figsize=(30,20), ax = ax4)


#mean taxi time per handler
ax5 = fig.add_subplot(525)
ax5.set_xlabel('Handler')
ax5.set_ylabel('Average Taxi Time per Handler (Minutes)')
ax5.set_title('Fig 5: Average Taxi Time per Handler (Minutes)') 
flightsSub.groupby('airline')['taxiTime'].agg(['mean']).sort_values('mean', ascending=False).plot(kind='bar',  figsize=(30,20), ax=ax5)

ax6 = fig.add_subplot(526)
#mean taxi time per handler
flightsSub.groupby('handler')['taxiTime'].agg(['mean']).sort_values('mean', ascending=False).plot(kind='bar',  figsize=(30,20), ax=ax6)

ax7 = fig.add_subplot(527)
#distribution of taxi time to help set theshold for investigations
flightsSub['taxiTime'].plot.hist(bins=242,xlim=(-40,200), ax=ax7)

#setting column names for aggregates, found here:  https://stackoverflow.com/questions/19078325/naming-returned-columns-in-pandas-aggregate-function
#because resetting index gave all columns heading of 'mean'
taxiTimeTemp = flightsSub.groupby(['airline']).agg(taxiMean=('taxiTime','mean')).reset_index()
rowCountTemp = flightsSub.groupby(['airline']).agg(numFlights=('rowCount','sum')).reset_index()
invThreshTemp = flightsSub.groupby(['airline']).agg(investigationMean=('investigationThreshold','mean')).reset_index()
flightsSubScatter = taxiTimeTemp.merge(rowCountTemp)
flightsSubScatter = flightsSubScatter.merge(invThreshTemp)

ax8 = fig.add_subplot(528)
#plot of % times airlines poass 15 minutre threshold taxitime
flightsSubScatter.groupby('airline')['investigationMean'].agg(['mean']).sort_values('mean', ascending=False).plot(kind='bar',  figsize=(30,20), ax = ax8)


#fig2 = plt.figure()
#ax9 = fig2.add_subplot(211)
plt.figure(figsize=(10,10))
sns.scatterplot(x='taxiMean',y='investigationMean', data=flightsSubScatter, hue='airline', size='numFlights',s=200,sizes=(200, 1000))
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

#sort by top 10 airlines and plot again
flightsSubScatter = flightsSubScatter.sort_values(by=['numFlights'],ascending=False)
flightsSubScatter = flightsSubScatter.head(10)
#flightsSubScatter = pd.concat([taxiTimeTemp, rowCountTemp.reindex(taxiTimeTemp.index), invThreshTemp.reindex(taxiTimeTemp.index)],axis=1)
plt.figure(figsize=(10,10))
#ax10 = fig2.add_subplot(212)
sns.scatterplot(x='taxiMean',y='investigationMean', data=flightsSubScatter, hue='airline', size='numFlights',s=200,sizes=(200, 1000))
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

#repeat scatter chart for handler
taxiTimeTemp = flightsSub.groupby(['handler']).agg(taxiMean=('taxiTime','mean')).reset_index()
rowCountTemp = flightsSub.groupby(['handler']).agg(numFlights=('rowCount','sum')).reset_index()
invThreshTemp = flightsSub.groupby(['handler']).agg(investigationMean=('investigationThreshold','mean')).reset_index()
flightsSubScatter = taxiTimeTemp.merge(rowCountTemp)
flightsSubScatter = flightsSubScatter.merge(invThreshTemp)

#fig2 = plt.figure()
#ax9 = fig2.add_subplot(211)
plt.figure(figsize=(10,10))
sns.scatterplot(x='taxiMean',y='investigationMean', data=flightsSubScatter, hue='handler', size='numFlights',s=200,sizes=(200, 1000))
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

```
